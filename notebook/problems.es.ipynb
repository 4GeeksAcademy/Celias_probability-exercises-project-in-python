## Ejercicio 1 

Dos dados se lanzan una vez y se observa el total obtenido. Usa una simulación para encontrar la probabilidad estimada de que la puntuación sea mayor a 7 o un número par. Una simulación es una repetición del mismo experimento multiples veces para observar su comportamiento:

- Ejecuta el experimento 1000 veces (lanza 2 dados 1000 veces, y suma el número de ambos dados).
- Lleva ua cuenta de los números y las veces que la suma fue mayor a 7 o un número par.
- Divide el número del paso 2 entre el número de iteraciones(1000).

```python
import random
import numpy as np
```

```python
# OPCIÓN 1 SIN BUCLE

dice_1 = np.random.randint(1,7, 1000)
dice_2 = np.random.randint(1,7,1000)
dice_sum = dice_1 + dice_2

recount = np.count_nonzero((dice_sum >= 7) | (dice_sum % 2 == 0)) # La barra vertical sirve como "OR"
dice_prob = recount/len(dice_1)

print(f"La probabilidad estimada de que la puntuación obtenida sea mayor a 7 o un número par es de {dice_prob}")
```

```python
#OPCIÓN 2 CON BUCLE

def dice_prob (rounds):
    
    recount = 0

    for i in range(rounds):

        dice_1 = np.random.randint(1,7)
        dice_2 = np.random.randint(1,7)
        dice_sum = dice_1 + dice_2

        if (dice_sum >= 7) or (dice_sum % 2 == 0):

            recount += 1
            prob = recount/rounds

    return print(f"La probabilidad estimada de que la puntuación obtenida sea mayor a 7 o un número par es de {prob}")


dice_prob(1000)
```

## Ejercicio 2

Una caja contiene 10 bolas blancas, 20 bolas rojas y 30 bolas verdes. Si tomamos 5 bolas de la caja reemplazándolas (tomando una bola, anotando el color y luego regresandola a la caja). Queremos saber la probabilidad de:

1. Tomar 3 blancas y 2 rojas.
2. Toas todas del mismo color.

Ejecuta el experimento 1000 veces y calcula las probabilidades que mencionadas.

```python
ball_box = {}

# Crea la cajas con las bolas

for i in range(60):
    if i < 10:
        ball_box[i] = "White"
    elif (i > 9) and (i < 30):
        ball_box[i] = "Red"
    else:
        ball_box[i] = "Green"

            
# Mi código ------------  Convertimos el diccinario "ball_box" en una lista de valores para poder iterar más fácilmente
ball_list = list(ball_box.values())

# Creamos los contadores para cada suceso
recount_3W_2R = 0
recount_same_color = 0
rounds = 1000

# Simulamos el proceso de coger 5 bolas e iteramos de manera que obtengamos un conteo de los colores resultantes en cada ronda.
for ball in range(rounds):
    recount_white = 0
    recount_red = 0
    recount_green = 0
    result = np.random.choice(ball_list, 5)

    if 'Red' in result:
        recount_red += np.sum(result == 'Red')

    if 'White' in result:
        recount_white += np.sum(result == 'White')
    
    if 'Green' in result:
        recount_green += np.sum (result =='Green')
        
    # Calculamos la probabilidad de tomar 3 blancas y 2 rojas
    if recount_white == 3 and recount_red == 2:
        recount_3W_2R += 1

    # Calculamos la probabilidad de tomar todas del mismo color
    if recount_white == 5 or  recount_red == 5 or recount_green == 5:
        recount_same_color += 1   
    
# Calculamos las probabilidades finales
prob_3W_2R= recount_3W_2R / rounds
prob_same_color = recount_same_color / rounds

print(f"Probabilidad de tomar 3 blancas y 2 rojas: {round(prob_3W_2R, 4)}")
print(f"Probabilidad de tomar todas del mismo color: {round(prob_same_color, 4)}")
```
